## 📈:In-Depth Financial and Tier Analysis – Executive Report


**Executive Summary**

Analysis of 3,000 retail banking customers reveals that mid-tier clients dominate by number, but higher-tier segments drive value. In particular, the Jade tier is the largest (1,331 customers) and, along with Silver (767), contributes the highest average account balances (mean deposits ~$696k for Jade, ~$658k for Silver). Platinum clients (317 members) are smallest and, surprisingly, have lower average balances than Jade/Silver.
Deposits and transactional accounts are strongly linked: Bank deposits correlate very highly with checking account balances (r≈0.84) and strongly with savings (r≈0.75). This indicates depositors also maintain liquid checking accounts and consider savings central to their finances.
Credit and loan usage shows moderate coupling: Customers with higher credit card balances tend to have slightly higher loan and deposit balances. Notably, bank loans correlate moderately with business lending (r≈0.42), suggesting business clients use both personal and business credit.
Income is a weak segmentation factor: Estimated income has only weak-to-moderate correlation with product use (max ~0.37 with superannuation or lending). This implies behavior-driven segmentation (based on product usage patterns) is more meaningful than pure income tiers.
Strategic focus: Leverage core product synergies (deposits/checking/savings), cross-sell credit and lending products, and align loyalty tiers with value-driving behavior. For example, Jade/Silver customers (with high deposit use) should be prioritized, while Platinum tier structures may be revised to reflect actual wallet share.


**Client Profile Overview**

Demographics and Nationality: The customer base is majority European (1,309 clients), followed by Asian (754) and American (507) nationals. Other nationalities (Australian, African) each account for <10%. A variety of occupations are represented, but analysis focuses on banking behavior rather than specific job roles.

Fee Structure: Most clients are on the High fee banking plan (1,476 customers), with Mid (962) and Low (562) fees following. This skew toward higher fee structures suggests many clients maintain substantial account activity.

Loyalty Tiers: Jade is the largest loyalty classification (1,331 customers), followed by Silver (767), Gold (585), and Platinum (317). This ordering is notable: Jade members outnumber Platinum and Gold combined. Loyalty tier counts will inform engagement strategies.

Income Bands: The majority of customers fall into the Medium income band (1,517 customers), with a sizable Low-income segment (1,027). Only 456 clients are in the High-income band. This distribution underscores that most clients are middle-earning, so tailoring products across income levels is important.

Other Profile Stats: Nearly two-thirds of clients hold at least one credit card (with 1 card: 1,922 clients; 2 cards: 765; 3 cards: 313), indicating moderate credit usage. Ownership of assets and risk profiles vary, but Key Baseline: high-fee and Jade-tier clients dominate, which aligns with high-account-balance behavior.

**Financial Behavior and Product Usage**

Core Deposit-Checking Relationship: Customers with larger bank deposit balances almost invariably have larger checking account balances. This “liquidity synergy” suggests customers treat checking accounts as a primary channel for spending/transfers while maintaining deposits. Savings accounts also track strongly with deposits (correlation ≈0.75), indicating that efforts to boost savings (e.g. via targeted offers) could lift overall deposit levels.

Savings vs. Checking: Savings and checking balances exhibit a moderate correlation (r≈0.46). This means while related, some customers prioritize one over the other. For example, depositors who hold a portion in savings may use checking for day-to-day needs, whereas others spread funds more evenly.

Credit and Loans: Credit usage is modestly linked to other products. Higher credit card balances correlate slightly with higher bank loans and deposits (r≈0.37–0.38). This could reflect customers funding spending or investments via both credit and deposits. Banks should note that credit card holders often maintain deposit accounts, presenting cross-sell opportunities. Separately, bank loans and business lending correlate (r≈0.42), signaling that business clients often take both personal and business loans. This suggests bank campaigns to cross-promote business lending to personal-loan customers (and vice versa) could be effective.

Product Distribution: Nearly all customers maintain deposit accounts (the basis of their banking relationship). Checking and savings are widely used among depositors. Foreign currency accounts (FCA) show weaker ties (r≈0.31 with savings), implying they serve niche international/business needs. Thus, FCAs are likely leveraged by corporate or overseas-oriented clients rather than ordinary savers.
Key Takeaway: Behavioral patterns drive product usage more than demographic labels. For instance, income alone shows only weak influence on saving or credit habits (lowest correlations ~0.26–0.30). Therefore, segmenting customers by how they use services (deposits vs. loans vs. credit) is more actionable than by income bracket.

**Loyalty Tier Segmentation**

Tier Population: Jade (1,331) and Silver (767) make up ~73% of all clients. Gold (585) and Platinum (317) are smaller groups. Jade and Silver customers are therefore the core segments by count.

Tier Value Alignment: Analysis of average balances by tier shows that Jade and Silver customers hold the highest assets. For example, Jade clients average $696k in deposits, with Gold/Silver close behind ($664k–$658k). In contrast, Platinum customers average only ~$612k. Thus, Jade and Silver segments outperform Platinum in raw balances.

Income by Tier: Silver members have the highest mean income ($175k), followed closely by Jade ($172k) and Gold ($168k). Platinum incomes are lowest ($164k). Notably, Platinum tier is not composed of the very richest customers – their wallets are smaller on average.

Implications: The loyalty program tiering is misaligned with value. Lower tiers (Jade/Silver) contain more high-value clients than the Platinum tier. This suggests a re-prioritization: treat Jade and Silver segments as premium in engagement and rewards, and reevaluate the Platinum criteria. For example, marketing offers traditionally reserved for Platinum clients might better boost retention/value among Jade members.

**Correlation Analysis**

The correlation heatmap of key numeric features (deposits, loans, accounts, etc.) reveals several actionable relationships:

Deposits ↔ Checking Accounts (r≈0.84): A very strong positive correlation. Customers with larger deposit balances almost always hold proportionately larger checking balances. Insight: Checking accounts serve as a liquidity hub for depositors. Strategic cross-sell (e.g. premium checking accounts) can strengthen deposit relationships.

Deposits ↔ Savings Accounts (r≈0.75): A strong link. High-deposit customers consistently use savings accounts as part of their portfolio. Implication: Promotions or benefits for savings accounts will likely boost overall deposits. Bundling savings products with deposit accounts could deepen engagement.

Checking ↔ Savings Accounts (r≈0.46): A moderate correlation. Some customers prioritize one over the other; not all depositors maintain high balances in both. This suggests different liquidity preferences within the customer base.

Credit Card Balances ↔ Loans/Deposits (~r=0.37–0.38): Modest positive correlations. Customers with higher credit card debt tend also to have somewhat higher loan amounts and deposits. It indicates that credit usage often coexists with active banking relationships (possibly through financing business or spending needs). Opportunity: Customers with significant credit usage may be receptive to loan or deposit offers, and vice versa.

Bank Loans ↔ Business Lending (r≈0.42): A moderate correlation. This signals that clients taking personal loans are also likely to utilize business loans. Opportunity: Cross-selling business lending products to personal-loan customers (and the reverse) can increase share of wallet.

Income ↔ Products: The highest income-product correlations are only moderate (e.g. 0.37 with superannuation savings, 0.33 with business lending). The lowest are weak (e.g. 0.26 with savings accounts, 0.30 with credit balances). Key insight: Income is not a dominant predictor of product engagement. High-income clients only slightly favor retirement/business products. Instead, product usage patterns (behavior) are more telling. This justifies focusing segmentation on usage (e.g. “deposit-rich vs. loan-rich” customers) rather than just income brackets.

Foreign Currency Account (FCA) Ties: FCA balances correlate moderately with bank deposits (r≈0.41) and business lending (r≈0.37), but weakly with savings. This suggests FCAs are primarily held by clients (often business or international) who also keep large cash/deposit balances. FCAs are less relevant to typical retail savers.

**Strategic Insights and Recommendations**

Focus on Core Deposit Products: Given the deep interconnection of deposits, checking, and savings, bundle and promote these products together. For example, reward customers who aggregate deposits into both checking and savings accounts. Use the strong deposit-check linkage to upsell premium checking accounts to high-deposit clients.
Behavior-Driven Segmentation: Emphasize customer behavior profiles (e.g. “high-deposit savers”, “active borrowers”, “credit card payers”) over pure income or demographic segments. Marketing and loyalty programs should target how clients use products. For instance, launch campaigns for “deposit maximizers” highlighting multi-account benefits, rather than targeting by income bracket.
Cross-Selling Opportunities: Leverage the observed product correlations:

Deposit-Savings: Encourage customers with large deposits to open/use savings accounts via incentives (higher interest, bundled services).
Credit-Loans: Target credit-active customers with personalized loan offers (personal, auto, business), since credit use often co-occurs with borrowing. Similarly, promote credit cards to loyal depositors who show capacity (e.g. high net balances) to spend.
Business vs. Personal Products: Identify clients with business lending (or vice versa) and market complementary loan products, as many already engage both.

Reassess Loyalty Tier Benefits: Align rewards with customer value. Since Jade and Silver tiers hold the most assets, consider upscaling benefits for these groups (even if they are “lower” tiers) and revisiting Platinum criteria. For example, introduce tiered loyalty incentives based on account balances or activity, ensuring high-asset Jade customers receive platinum-level perks.
Leverage Foreign Currency Accounts: Market FCAs to international or corporate clients who need multi-currency liquidity, as indicated by their association with high deposits/business lending. Do not prioritize FCAs for domestic savers.

**Final Conclusion**

This comprehensive analysis highlights that client behavior is key in driving banking value. Jade and Silver clients — not just the top-tier Platinum — are the bank’s most valuable segments, given their high deposit and income metrics. Core banking products (deposits, checking, savings) are deeply intertwined, so integrated offerings around these should be the strategic centerpiece. Meanwhile, income-based targeting is less effective — product usage patterns (and loyalty tier behavior) should guide segmentation and marketing. Key Takeaways: Jade/Silver customers are high-value and should be the focus of premium retention efforts. Depositors also use checking and savings extensively; cross-selling and bundling here can deepen relationships. Credit card holders often overlap with loan customers, revealing upsell paths. Finally, segment and reward customers based on how they use products, not just how much they earn. This behavior-driven approach will maximize customer lifetime value and align bank resources with the most profitable relationships.
